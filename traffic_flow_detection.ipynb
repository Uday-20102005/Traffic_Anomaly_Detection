```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
```

```python
df=pd.read_csv("embedded_system_network_security_dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python

df['tcp_flags_SYN-ACK'].value_counts()
```

```python
features=df.drop('label',axis=1)
for col in df.columns:
    if df[col].dtype == 'bool':
        df[col] = df[col].astype(int)
```

```python
features.columns
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaled_features=scaler.fit_transform(features)
```

```python
scaled_features_df=pd.DataFrame(scaled_features,columns=features.columns)
scaled_features_df.head()
```

```python
n_estimators=100
contamination=0.04
sample_size=256
```

```python
model=IsolationForest(n_estimators=n_estimators,contamination=contamination,
                      max_samples=sample_size,random_state=42)

model.fit(scaled_features_df)
anomaly_lables=model.predict(scaled_features_df)
scaled_features_df['anomaly']=anomaly_lables
```

```python
scaled_features_df['anomaly'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=scaled_features_df,x='packet_size',y='packet_count_5s',hue='anomaly')
plt.tight_layout()
plt.show()
```
